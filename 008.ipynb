```python
from collections import defaultdict

def parse_data(data: str):
    grid = defaultdict(tuple)
    lines = data.split("\n")
    for i, line in enumerate(lines):
        for j, c in enumerate(line):
            if c != ".":
                grid[(i,j)] = c

    return grid
```

```python
from collections import defaultdict

def parse_data_into_map(data: str):
    grid = defaultdict(str)
    lines = data.split("\n")
    for i, line in enumerate(lines):
        for j, c in enumerate(line):
            grid[(i,j)] = c

    return grid
```

```python
def reverse_dict(d: defaultdict):
    rev = defaultdict(list)
    for k, v in d.items():
        rev[v].append(k)
    return rev
```

```python
from itertools import combinations

def get_combinations(lst: list, length: int = 2) -> list:
    return list(combinations(lst, length))
```

```python
from typing import List

def calculate_antinodes(nodes: List[tuple]):
    t1, t2 = nodes
    distance_vector = (t2[0] - t1[0], t2[1] - t1[1])
    return [(t1[0]-distance_vector[0], t1[1]-distance_vector[1]), (t2[0]+distance_vector[0], t2[1]+distance_vector[1])]
```

```python
from typing import List

def calculate_all_antinodes(nodes: List[tuple], data_map: Dict[Tuple[int, int], str]):
    t1, t2 = nodes
    distance_vector = (t2[0] - t1[0], t2[1] - t1[1])
    lst = [t1, t2]

    # calculate backwards
    while data_map[(t1[0]-distance_vector[0], t1[1]-distance_vector[1])] != '':
        t1 = (t1[0]-distance_vector[0], t1[1]-distance_vector[1])
        lst.append(t1)

    # calculate forwards
    while data_map[(t2[0]+distance_vector[0], t2[1]+distance_vector[1])] != '':
        t2 = (t2[0]+distance_vector[0], t2[1]+distance_vector[1])
        lst.append(t2) 


    return lst
```
