## Part ONE

```python
data_test = """MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX"""
```

```python
import re
```

```python
# regex pattern to match string "XMAS"
pattern = r'XMAS'
pattern_reverse = r'SAMX'
```

```python
# read data from file
data = ''
with open('./data/004.txt') as f:
    for line in f:
        data += line
```

```python
data
```

```python
data_list = data_test.split('\n')
data_list = data.split('\n')
```

```python
#count of XMAS in each string
def count_xmas(data_list):
    return sum(len(re.findall(pattern, line)) for line in data_list)
```

```python
def count_samx(data_list):
    return sum(len(re.findall(pattern_reverse, line)) for line in data_list)
```

```python
# transpose list of strings 90 degrees
def transpose(data_list):
    return [''.join([data_list[j][i] for j in range(len(data_list))]) for i in range(len(data_list[0]))]
```

```python
# Function to extract all diagonals
def get_diagonals(matrix):
    size = len(matrix)
    diagonals = []

    # Top-left to bottom-right diagonals
    for d in range(2 * size - 1):
        diagonal = []
        for i in range(size):
            j = d - i
            if 0 <= j < size:
                diagonal.append(matrix[i][j])
        diagonals.append("".join(diagonal))

    # Top-right to bottom-left diagonals
    for d in range(2 * size - 1):
        diagonal = []
        for i in range(size):
            j = i - (size - 1 - d)
            if 0 <= j < size:
                diagonal.append(matrix[i][j])
        diagonals.append("".join(diagonal))

    return diagonals
```

```python
get_diagonals([list(x) for x in data_list])
```

```python
res = 0
res += count_xmas(data_list)
res += count_samx(data_list)
res += count_xmas(transpose(data_list))
res += count_samx(transpose(data_list))
res += count_xmas(get_diagonals([list(x) for x in data_list]))
res += count_samx(get_diagonals([list(x) for x in data_list]))
```

```python
#change list of stings into aray of characters
matrix = [list(x) for x in data_list]
```

```python
def is_first_diagonal(i,j):
    if matrix[i-1][j-1] == 'M' and matrix[i+1][j+1] == 'S':
        return True
    if matrix[i-1][j-1] == 'S' and matrix[i+1][j+1] == 'M':
        return True
    return False
```

```python
def is_second_diagonal(i,j):
    if matrix[i-1][j+1] == 'M' and matrix[i+1][j-1] == 'S':
        return True
    if matrix[i-1][j+1] == 'S' and matrix[i+1][j-1] == 'M':
        return True
    return False
```

```python
# iterate over all characters in matrix
res = 0

for i in range(1,len(matrix) - 1):
    for j in range(1, len(matrix[0]) - 1):
        if matrix[i][j] == 'A':
            if is_first_diagonal(i,j) and is_second_diagonal(i,j):
                res += 1
```
